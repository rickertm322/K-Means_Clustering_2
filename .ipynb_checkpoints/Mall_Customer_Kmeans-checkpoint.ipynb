```python
df.corr(numeric_only = True)
#pandas.dataframe.corr() gives the pairwise correlation of each two columns 
#of the dataframe. Default is the Pearson correlation coefficient (described
#later). Original code did not specify numeric_only, so needed to debug to only
#take into account numeric data. 

#Pearson correlation coefficient (PCC):
#Let cov(x,y) be the covariance of x and y, and stdx, stdy to be the standard
#deviations of x and y respectively. Let p be the PCC
#p = cov(x,y)/(stdx*stdy)
#Covariance represents the tendency of the data to change together (x and y
#have some sort of relation). 
#The PCC is a normalized number which represents the tendency of the data to 
#have a linear correlation. 
# -1<p<0 indicates a negative linear relationship
# p = 0 (or very small) indicates no linear relationship
# 0<p<1 indicates a positive linear relationship
# p values of 1 and -1 indicate perfect linear correlation as well. For most 
#data we expect a value somewhere in between. 

#Analysis of the data below:
#Age and annual income have a low correlation factor, so they likely aren't 
#very relation. However age and spending score have a higher PCC, so we can say
#that as age goes up, generally people spend less. 
#Spending score and annual income also don't seem to have a correlation,
#as the PCC is very low. 
```

## Data Labeling

```python
encoder=LabelBinarizer()
#sklearn.preprocessing.LabelBinarizer() transforms a multi-class labels into a 
#binary system (literally 1, 0) of belonging to a class or not. 
df['Gender']=encoder.fit_transform(df['Gender'])
#Transforms the 'Gender' column into a binary (1,0) classification; 1 = male
#2 = female

#print(df['Gender']) #using to visualize the change in data
```

```python
skew(df)
#measures the skew on a set of data; if the data is normally/uniformly 
#distributed, the skew will be ~ 0. A positive skew indicates the data has 
#more weight on the right, and a negative skew indicates the data has more 
#weight on the left. 

#Gender has a higher skew to the right (male), as does age (unsuprising) and
# income. However, spending score has a skew to the left (smaller). 
```

```python
kurtosis(df)
#Measures the kurtosis of the data, or the 'tailedness'. Tailedness is an 
# informal term to describe how much the data stretches beyond the central 
# mass of data within a normally distributed graph. 
#There are multiple interpretations of kurtosis; automatic is Fisher. In this:
    #Normal distribution = 0 kurtosis, typical bell curve looking thing
    #Thin-tailed < 0 kurtosis, has little data outside of center
    #Thick tailed > 0 kurtosis, has lots of data outside of center. 
#For this data, we see the gender has a very small kurtosis (expected, since 
# it is binary). The rest have small kurtosis as well.
```

## Plot Elbow Method For Choosing Value Of K

```python
cluster_no=[i for i in range(1,12)]
print(cluster_no)
WCSS=[]
for i in cluster_no:
    cluster=KMeans(init='k-means++',n_clusters=i,random_state=42)
    cluster.fit(df)
    WCSS.append(cluster.inertia_)
```

```python
plt.plot(cluster_no,WCSS,marker='o')
plt.xlabel('Cluster Numbers')
plt.ylabel('WCSS')
```

## K-Means Model

```python
cluster=KMeans(init='k-means++',n_clusters=4,random_state=42)
df['Cluster_No']=cluster.fit_predict(df)
```

```python
cluster.cluster_centers_
```

# Visualize Cluster3

```python
# Creating figure
plt.style.use('dark_background')
fig = plt.figure(figsize = (20, 15))
ax = plt.axes(projection ="3d")
# Creating plot
ax.scatter3D(df['Age'][df['Cluster_No']==0], df['Annual_Income'][df['Cluster_No']==0], df['Spending_Score(1-100)'][df['Cluster_No']==0], c = "red", s=150)
ax.scatter3D(df['Age'][df['Cluster_No']==1], df['Annual_Income'][df['Cluster_No']==1], df['Spending_Score(1-100)'][df['Cluster_No']==1], c = "green", s=150)
ax.scatter3D(df['Age'][df['Cluster_No']==2], df['Annual_Income'][df['Cluster_No']==2], df['Spending_Score(1-100)'][df['Cluster_No']==2], c = "yellow", s=150)
ax.scatter3D(df['Age'][df['Cluster_No']==3], df['Annual_Income'][df['Cluster_No']==3], df['Spending_Score(1-100)'][df['Cluster_No']==3], c = "blue", s=150)
ax.view_init(10, 170)
plt.title("3D scatter plot")
plt.show()
```
